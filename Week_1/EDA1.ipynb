```python
def replace_missing_with_median(dataframe, column_name):
    median_value = dataframe[column_name].median()
    updated_dataframe = dataframe.copy()
    updated_dataframe[column_name].fillna(median_value, inplace=True)
    return updated_dataframe

def replace_missing_with_median_for_all_columns(dataframe):
    updated_dataframe = dataframe.copy()
    for column in updated_dataframe.columns:
        if updated_dataframe[column].isnull().any():
            updated_dataframe = replace_missing_with_median(updated_dataframe, column)
    return updated_dataframe

df = replace_missing_with_median_for_all_columns(df)
```
