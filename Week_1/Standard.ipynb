```python
import pandas as pd
# Load the dataset
df = pd.read_csv('E:\P288_Clustering_Project\P288_Group1_Clustering_Project_Excelr\Dataset\IQR.csv')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming 'Country' is the categorical column
numerical_cols = df.columns[df.dtypes != 'object'].values

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[numerical_cols])

"""# Perform KMeans clustering
kmeans = KMeans(n_clusters=5)
clusters = kmeans.fit_predict(df_scaled)"""
agg = AgglomerativeClustering(n_clusters=3)
labels_agg = agg.fit_predict(df_scaled)

# Add the cluster labels to the dataframe as a new feature
df['Cluster'] = clusters

print(df.head())
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn import metrics

# Assuming 'df_normalized' is your DataFrame

# K-Means
kmeans = KMeans(n_clusters=5)
labels_kmeans = kmeans.fit_predict(df_scaled)

# Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=3)
labels_agg = agg.fit_predict(df_scaled)

# DBSCAN
dbscan = DBSCAN(eps=0.3)
labels_dbscan = dbscan.fit_predict(df_scaled)

# Calculate metrics
for name, labels in zip(['K-Means', 'Agglomerative', 'DBSCAN'], [labels_kmeans, labels_agg, labels_dbscan]):
    print(f"\n{name}:")
    print(f"Silhouette Score: {metrics.silhouette_score(df_scaled, labels)}")
    print(f"Calinski-Harabasz Index: {metrics.calinski_harabasz_score(df_scaled, labels)}")
    print(f"Davies-Bouldin Index: {metrics.davies_bouldin_score(df_scaled, labels)}")
```
